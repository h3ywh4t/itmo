**Условие:** Для каждого варианта требуется построить линейную модель (предполагая нормальность распределения ошибок, некоррелированность компонент, гомоскедастичность), вычислить
оценки коэффициентов модели и остаточной дисперсии, построить для них доверительные интервалы, вычислить коэффициент детерминации, проверить указанные в условии гипотезы с помощью построенной линейной модели. Из встроенных функций разрешается пользоваться квантильными функциями и средствами для квадратичной оптимизации.

1. Постройте линейную модель, где в качестве независимых переменных выступают продолжительность,
"танцевальность" и энергичность (вместе со свободным коэффициентом), зависимой - популярность.
2. Проверьте следующие подозрения:
- Чем больше энергичность, тем больше популярность
- Популярность зависит от продолжительности
- Проверьте гипотезу Но о равенстве одновременно нулю коэффициентов при энергичности и "танцевальности" против альтернативы $H_1$ = $H_0$


## **Решение**

- **Линейная регрессия**. Модель: \$y = X\beta + \varepsilon\$, где \$X\$ — матрица регрессоров (содержащая столбец единиц для свободного члена). Оценки МНК-коэффициентов находятся решением нормальных уравнений:

  $$
  \hat\beta = (X^T X)^{-1}X^T y.
  $$

  Замкнутая форма решения МНК.

  Формула \$\hat\beta=(X^TX)^{-1}X^T y\$ даёт наименьшее \$\sum e\_i^2\$. Здесь \$X\$ включает столбец единиц (для свободного члена) и столбцы предикторов (duration, danceability, energy). Компоненты \$\beta\_1,\beta\_2,\beta\_3\$ показывают изменение популярности при изменении соответствующих регрессоров, при фиксированных остальных факторах.

- **Оценка дисперсии ошибок**: для полученной модели вычисляем остатки \$e=y-\hat y\$ и

  $$
    \hat\sigma^2 = \frac{1}{n - k}\sum_{i=1}^n e_i^2,
  $$

  где \$k\$ — число параметров (включая свободный член).


- **Стандартная ошибка коэффициента** \$\hat\beta\_j\$:

  $$
    \mathrm{Var}(\hat\beta) = \hat\sigma^2 (X^TX)^{-1},
    \qquad
    \mathrm{SE}(\hat\beta_j)=\sqrt{\hat\sigma^2\,(X^TX)^{-1}_{jj}}.
  $$

* **t-статистика для проверки \$\beta\_j\$:**

  $$
    t = \frac{\hat\beta_j - \beta_{j0}}{\mathrm{SE}(\hat\beta_j)},
  $$

  где \$\beta\_{j0}\$ — значение коэффициента при \$H\_0\$ (обычно 0). Для двустороннего теста \$p\$-значение: \$p = 2(1 - F\_{t}(|t|; df))\$, где \$df = n - k\$. Если тест односторонний (например \$\beta\_j>0\$), то \$p = 1 - F\_t(t;df)\$ при \$t>0\$.

После оценки \$\hat\beta\$ вычисляется несмещённая оценка дисперсии ошибок \$\hat\sigma^2\$. Стандартная ошибка \$\mathrm{SE}(\hat\beta\_j)=\sqrt{\hat\sigma^2 (X^TX)^{-1}\_{jj}}\$ отражает неопределённость оценки. При гипотезе \$H\_0: \beta\_j=0\$ статистика \$t = \hat\beta\_j/\mathrm{SE}(\hat\beta\_j)\$ имеет приближенно \$t\$-распределение с \$df=n-k\$. Мы сравниваем значение \$t\$ с критическими значениями или вычисляем \$p\$-значение. Для двусторонних тестов учитываем обе стороны критической области; для одностороннего \$\beta\_3>0\$ интересует только правая сторона.

**F-статистика для совместной проверки (несколько ограничений)**: если проверяется \$q\$ ограничений (например, \$\beta\_2=0\$ и \$\beta\_3=0\$, \$q=2\$), то сравнивают значение функции качества (SSE) полной и ограниченной моделей. F-статистика:

  $$
    F = \frac{(\mathrm{SSE}_{\text{огр}} - \mathrm{SSE}_{\text{пол}})/q}{\mathrm{SSE}_{\text{пол}}/(n - k)},
  $$

  где \$\mathrm{SSE}\$ — сумма квадратов остатков. Под \$H\_0\$ \$F\$ распределён \$F(q, n-k)\$.

При совместной проверке \$\beta\_2=\beta\_3=0\$ сравниваем полную модель (с danceability и energy) и ограниченную (без них, оставляя только duration). Если в ограниченной модели существенно возрастает сумма квадратов остатков, то \$F\$-статистика будет большой и гипотезу \$H\_0\$ отвергают.

## **Решение**

* **Составление матрицы данных.** Пусть \$y\_i\$ — популярность \$i\$-й песни, а \$x\_{i1}=\$ duration (мс), \$x\_{i2}=\$ danceability, \$x\_{i3}=\$ energy. Строим матрицу \$X\$ размера \$n\times 4\$ с первым столбцом из единиц (для \$\beta\_0\$) и тремя столбцами предикторов.

* **Нахождение коэффициентов.** Считаем оценки по формуле: \$\hat\beta=(X^TX)^{-1}X^Ty\$. В результате получаем (примерные значения):

  $$
    \hat\beta_0\approx 44.61,\quad \hat\beta_{\text{duration}}\approx -2.85\times10^{-6},\quad
    \hat\beta_{\text{dance}} \approx 14.48,\quad \hat\beta_{\text{energy}}\approx -0.257.
  $$

  Это означает, что при прочих равных величинах продолжительности, танцевальности и энергичности популярность примерно равна \$44.6 + (-2.85\cdot10^{-6})\cdot \text{duration} + 14.48\cdot \text{danceability} -0.257\cdot\text{energy}\$.

* **Вычисление остаточной дисперсии.** Остатки \$e\_i = y\_i - \hat y\_i\$ дают сумму квадратов \$\mathrm{SSE}\_{\text{full}}\$. Оценка дисперсии ошибок \$\hat\sigma^2 = \mathrm{SSE}/(n - k)\$, здесь \$n\approx18835\$, \$k=4\$.

* **Стандартные ошибки и t-статистики.** Из \$\mathrm{Var}(\hat\beta)=\hat\sigma^2(X^TX)^{-1}\$ извлекаем стандартные ошибки:
  примерно

  $$
    \mathrm{SE}(\hat\beta_0)\approx 1.00,\quad
    \mathrm{SE}(\hat\beta_{\text{duration}})\approx 2.68\times10^{-6},\quad
    \mathrm{SE}(\hat\beta_{\text{dance}})\approx 1.02,\quad
    \mathrm{SE}(\hat\beta_{\text{energy}})\approx 0.746.
  $$

  Тогда \$t\$-статистики:

  $$
    t_0\approx 44.4,\quad
    t_{\text{duration}}\approx -1.064,\quad
    t_{\text{dance}}\approx 14.19,\quad
    t_{\text{energy}}\approx -0.344.
  $$

  Соответствующие \$p\$-значения (двусторонние) примерно:

  $$
    p_0\approx 0,\quad p_{\text{duration}}\approx 0.287,\quad
    p_{\text{dance}}\approx 0,\quad p_{\text{energy}}\approx 0.731.
  $$

  Интерпретация:

  * Перехват \$\beta\_0\$ значим (\$t\approx44.4\$, \$p<10^{-16}\$).
  * Коэффициент при \$duration\$ (\$\beta\_1\$): \$t=-1.064\$, \$p\approx0.287\$. *Не является статистически значимым* (не отвергаем \$H\_0:\beta\_1=0\$).
  * Коэффициент при \$danceability\$ (\$\beta\_2\$): \$t\approx14.2\$, \$p\approx0\$ — высокозначимый (отвергаем \$H\_0:\beta\_2=0\$).
  * Коэффициент при \$energy\$ (\$\beta\_3\$): \$t=-0.344\$, \$p\approx0.731\$. Не значим, фактически \$\hat\beta\_3\$ отрицательное.

## **Проверка гипотез**

  1. *Энергичность vs популярность*: По условию \$H\_0:\beta\_3\le0\$ против \$H\_1:\beta\_3>0\$. Поскольку \$\hat\beta\_3<0\$ и \$t\approx -0.344\$ (правее критического \$t\_{0.95}(18831)\approx1.645\$), мы не видим доказательств \$\beta\_3>0\$. Односторонний \$p\$-value \$\approx 0.63\$. **Гипотеза зависимо от энергичности НЕ подтверждается**.
  2. *Влияние продолжительности*: \$H\_0:\beta\_1=0\$ против двустороннего \$H\_1\$. Полученное \$t=-1.064\$, \$p\approx0.287>0.05\$. **Не отвергаем \$H\_0\$:** нет статистически значимой связи с продолжительностью.
  3. *Совместная проверка \$\beta\_2=\beta\_3=0\$.* Оцениваем две модели: полную (с danceability и energy) и ограниченную (только intercept и duration). Вычисляем SSE для каждой. Расчёт дает приближенно:

  $$
    \mathrm{SSE}_{\text{full}}\approx 8.939\times10^6,\quad
    \mathrm{SSE}_{\text{restricted}}\approx 9.034\times10^6.
  $$

  Разность SSE \$\approx 95703\$ на \$q=2\$ степенях свободы даёт \$F \approx 100.8\$. Степени свободы: \$(df\_1=2,;df\_2=n-4\approx18831)\$. \$p\$-значение F-теста оказалось \$\approx1.1\times10^{-16}\$ (очень малое). **Отвергаем \$H\_0\$** на любом разумном уровне значимости: совместно коэффициенты при \$danceability\$ и \$energy\$ не равны нулю (причём разница дала статистику почти исключительно за счёт \$danceability\$, так как \$\beta\_2\$ значим).

**Вывод:** Данные не подтверждают предположение «чем больше энергичность, тем выше популярность»: коэффициент \$\beta\_3\$ оказался отрицательным и статистически незначимым (\$p\approx0.73\$). **Популярность не зависит от энергичности** в этой модели. Аналогично, влияние продолжительности (\$\beta\_1\$) статистически незначимо (\$p\approx0.29\$): нет доказательств, что более длительные треки популярнее. Единственным значимым фактором оказалась танцевальность (\$\beta\_2>0\$, \$p\approx0\$). Совместный тест \$H\_0: \beta\_2=\beta\_3=0\$ привёл к значимому отклонению (очень малое \$p\$): фактически модель без этих факторов хуже описывает данные, но это за счёт \$\beta\_2\$. Таким образом, **вывод:** популярность песен существенно возрастает с увеличением танцевальности, а связи с длительностью или энергией не обнаружено.

Таким образом, результат регрессии: коэффициент при **танцевальности** значимо положителен, при прочих равных популярность растёт с увеличением danceability. Коэффициенты при длительности и энергичности статистически незначимы, поэтому гипотезы о зависимости от них отвергнуты.

```python
import numpy as np
from math import sqrt
from scipy import stats

import csv
data = []
with open('/mnt/data/song_data.csv', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        if row['song_popularity']=='' or row['song_duration_ms']=='' or row['danceability']=='' or row['energy']=='':
            continue
        y = float(row['song_popularity'])
        x1 = float(row['song_duration_ms'])
        x2 = float(row['danceability'])
        x3 = float(row['energy'])
        data.append([1.0, x1, x2, x3, y])

data = np.array(data)
X = data[:,0:4]
y = data[:,4]
n = len(y)
XtX = X.T @ X
XtY = X.T @ y
beta_hat = np.linalg.inv(XtX) @ XtY
y_pred = X @ beta_hat
resid = y - y_pred
SSE = np.sum(resid**2)
df = n - X.shape[1]
sigma2 = SSE/df

cov_beta = sigma2 * np.linalg.inv(XtX)
SE = np.sqrt(np.diag(cov_beta))
t_stats = beta_hat/SE
p_values = [2*(1 - stats.t.cdf(abs(t), df)) for t in t_stats]

print("=== Регрессия ===")
print("Коэффициенты β:", beta_hat)
print("SE:", SE)
print("t:", t_stats)
print("p-значения (двусторонние):", p_values)

t_energy = t_stats[3]
p_one_sided = 1 - stats.t.cdf(t_energy, df) if t_energy > 0 else stats.t.cdf(t_energy, df)
print("Односторонний p для β_energy>0:", p_one_sided)

X_restricted = X[:, [0,1]]
beta_res = np.linalg.inv(X_restricted.T @ X_restricted) @ (X_restricted.T @ y)
y_pred_res = X_restricted @ beta_res
SSE_res = np.sum((y - y_pred_res)**2)
q = 2
F_stat = ((SSE_res - SSE)/q) / (SSE/df)
p_F = 1 - stats.f.cdf(F_stat, q, df)
print("F-test β2=β3=0: F =", F_stat, "p =", p_F)
```

# **Задание 2**

**Условие:** Требуется проверить гипотезу о равенстве средних на каждом уровне фактора с помоью модели однофакторного дисперсионного анализа.
В файле sex_bmi_snmokers.csv приведены
данные о некоторых привычках и физиологических/антропологических показателях. Фактор - курит/не курит.
Выходная переменная - индекс массы тела

## **Решение**

**ANOVA (однофакторный дисперсионный анализ)** для \$k\$ групп. Общее число наблюдений \$N=\sum\_i n\_i\$. Обозначим \$\bar y\_i\$ — среднее в i-й группе, \$\bar y\$ — общее среднее. Тогда:

  * Межгрупповая сумма квадратов:

    $$
      \mathrm{SSB} = \sum_{i=1}^k n_i(\bar y_i - \bar y)^2.
    $$
  * Внутригрупповая сумма:

    $$
      \mathrm{SSW} = \sum_{i=1}^k \sum_{j=1}^{n_i} (y_{ij} - \bar y_i)^2.
    $$
  * Общее \$SST = \mathrm{SSB} + \mathrm{SSW} = \sum\_{i,j} (y\_{ij}-\bar y)^2\$. Формулы SSB и SSW приведены, например, в учебнике.
  * F-статистика:

    $$
      F = \frac{\mathrm{SSB}/(k-1)}{\mathrm{SSW}/(N-k)},
    $$

    распределена \$F(k-1, N-k)\$ при \$H\_0\$.


Однофакторный ANOVA при двух группах эквивалентен двухвыборочному \$t\$-тесту. Группы «курящие» (\$i=1\$) и «не курящие» (\$i=2\$) с размерами \$n\_1,n\_2\$. Межгрупповая дисперсия SSB показывает, насколько средние групп отличаются, а внутригрупповая SSW — вариативность внутри каждой группы. F-статистика \$F=(\mathrm{SSB}/1)/(\mathrm{SSW}/(N-2))\$ проверяет равенство \$\bar y\_1\$ и \$\bar y\_2\$. Если \$F\$ мало, гипотеза равенства средних принимается; если \$F\$ велико и \$p\<F\_{\alpha}(1,N-2)\$, \$H\_0\$ отвергают.

* **Результаты:**

  * Средние BMI: \$\bar y\_{\text{yes}}\approx30.708\$, \$\bar y\_{\text{no}}\approx30.652\$.
  * Межгрупповая SSB≈0.6993, внутригрупповая SSW≈49719.50.
  * \$F\approx0.01879\$, \$p\approx0.891\$.

При уровне \$\alpha=0.05\$ значение \$F=0.0188\$ дал \$p\approx0.891>0.05\$. Значимой разницы между средними BMI курящих и некурящих не обнаружено. **Не отвергаем \$H\_0\$**: средние BMI статистически равны для двух групп.

**Вывод:** Между средними BMI курящих и некурящих **нет значимой разницы** (\$p\approx0.891\$). Следовательно, **гипотеза равенства средних BMI принята**: курящий статус не влияет на средний индекс массы тела в этих данных.

## **Код**

```python
import pandas as pd
df_bmi = pd.read_csv('/mnt/data/sex_bmi_smokers.csv')
groups = df_bmi.groupby('smoker')['bmi']
means = groups.mean()
counts = groups.count()
grand_mean = df_bmi['bmi'].mean()
SSB = sum(counts[i] * (means[i] - grand_mean)**2 for i in means.index)
SSW = sum(((df_bmi[df_bmi['smoker']==i]['bmi'] - means[i])**2).sum() for i in means.index)
F_anova = (SSB/(len(means)-1)) / (SSW/(len(df_bmi)-len(means)))
p_anova = 1 - stats.f.cdf(F_anova, len(means)-1, len(df_bmi)-len(means))
print("=== ANOVA ===")
print("SSB =", SSB, "SSW =", SSW, "F =", F_anova, "p =", p_anova)
```
